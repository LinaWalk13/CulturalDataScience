## Music sales challenge

Write a script that:

1. Combines the tables of best-selling physical singles and best-selling digital singles on the Wikipedia page "List_of_best-selling_singles"
2. Adds a column which marks whether each row is from the list of physical singles or digital singles
3. Outputs the artist and single name for the year you were born. If there is no entry for that year, take the closest year after you were born.
4. Outputs the artist and single name for the year you were 15 years old.

```python
# Starter code...

%pip install pandas lxml beautifulsoup4 html5lib
import pandas as pd
```

```python
rawdata = pd.read_html("https://en.wikipedia.org/wiki/List_of_best-selling_singles")

digital = rawdata[3]
physical = rawdata[0]

digital['Type']= 'digital'
physical['Type']= 'physical'
print(digital)
print(physical)
df_combined = pd.concat([digital, physical])
```

```python
# finging songs closest to my birthday
count = 0
for index, row in df_combined.iterrows():
    if row['Released'] == 2004:  
        print(row['Single'])  
        count = count + 1

if count == 0: 
    df_combined['Difference'] = abs(df_combined['Released']-2004)
    closest_number = df_combined['Difference'].idxmin()
    for index, row in df_combined.iterrows():
        if row['Difference'] == closest_number:  
            print(row['Single'])  
            print(row['Released'])
```

```python
# finding songs fro years 2019 and 2020
for index, row in df_combined.iterrows():
    if row['Released'] == 2019 or row['Released'] == 2020:  
        print(row['Single'])  
```

## Space challenge

1. Make a single dataframe that combines the space missions from the 1950's to the 2020's
2. Write a script that returns the year with the most launches
3. Write a script that returns the most common month for launches
4. Write a script that ranks the months from most launches to fewest launches


```python
rawdata = pd.read_html("https://en.wikipedia.org/wiki/Timeline_of_Solar_System_exploration")

# connect dfs
df_combined = pd.concat(rawdata[0:9], ignore_index=True)

# remove missing value cases in columns of interest
df_cleaned = df_combined.dropna(subset=['Mission name', 'Launch date'])
# get years and get max from the newly defined 
years = df_cleaned['Launch date'].str[-5:]
pd.Series(years).value_counts().idxmax()

# gte needed package
from collections import Counter
month_count = Counter()
months = ['January', 'February', 'March', 'April', 'May', 'June', 
          'July', 'August', 'September', 'October', 'November', 'December']


# Loop through months
for month in months:
    # check occurence in column
    for date in df_cleaned['Launch date']: 
            if month in date: 
                month_count[month] += 1

# we sort them in order
sorted_months = sorted(month_count.items(), key=lambda x: x[1], reverse=True) # sorting by count in the tuples
print(sorted_months)
```
